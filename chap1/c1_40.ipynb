`add`と`cube`と`square`があれば`cubic`は定義できる．

```python
def cubic(a, b, c):
    return lambda x: x**3 + a*(x**2) + b*x + c
```

$x^3 - x^2 - 2x = x(x+1)(x-2)$を試しに解いてみる．

```python
my_cubic = cubic(-1, -2, 0)
x = np.linspace(-1.1, 2.1, 1000)
y = [my_cubic(i) for i in x]
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.axhline(lw=0.5, ls='--', c='red')
plt.show()
```
